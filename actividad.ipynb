### Cómo Leer y mostrar una imágen 

Tenemos una imagen llamada "ejemplo.jpg" en el mismo directorio de nuestro código, usaremos utilizar el módulo io de scikit-image para leer y pyplot de matplotlib para mostrar esta imagen, en codigo de python los importaremos asi:

```python
from skimage import io
import matplotlib.pyplot as plt
```

### Escala de grises

El modulo color de la libreria skimage, contiene la funcion rgb2gray la cual permite mapear  el espacio de color de 3 canales(RGB) en un espacio de escala de grises de 1 canal. a cada color se le asigna una valor entre 0-1 donde 1 es blanco y 0 negro. 

```python
from skimage import color
import matplotlib.pyplot as plt
img_gris = color.rgb2gray(img)
plt.imshow(img_gris, cmap='gray')
plt.show()
```

### Desarrollo del programa 

Teniendo las herramientas basicas para el procesamiento de imagenes en python, vamos a implimentar un programa con el que podamos separa las vias presentes en una imagen satelital y un del resto de elementos presentes en ella. es importante tener en cuenta que es mas facil trabajar con la imagen cuando esta esta en escala de grises. 

Importacion del las librerias skimage y matplolib, lectura lectura la imagen y mapeo a escala de grises

```python
from skimage import io,color
import matplotlib.pyplot as plt
img =io.imread('ejemplo1.jpg')
cimg =io.imread('ejemplo1.jpg')
img_gris = color.rgb2gray(img)
img3 = color.rgb2gray(img)
info=img.shape
img2=img_gris
```

Los caminos en la imagen se encuentar en un rango aproximado de 0.5 y 0.95 en escala de grises, por lo cual usando ciclos for anidados  se recorre las dimensiones de la imagen, pixel a pixel, realizando la verificacion en la imagen en escala de grises, y realizado modificaciones en la imagen original y en la del mapeo, con el fin de separar estos 2 elementos.

```python
for i in range(info[0]):
    for j in range(info[1]):
        if img2[i][j]<=0.95  and img2[i][j]>=0.5:
            aux=img[i][j]
            aux[0]=255
            aux[1]=255
            aux[2]=255
            img2[i][j]=0
        else:
            img2[i][j]=1
```

Se grafican los resultado obtenidos usando comandos de las librerias skimage y matplolib 

```python
plt.subplot(221)
plt.imshow(img2, cmap='gray')
plt.title('CAMINOS')
plt.subplot(222)
plt.imshow(img)
plt.title('RESTO DE ELEMENTOS')
plt.subplot(223)
plt.imshow(cimg)
plt.title('IMG ORIGINAL')
plt.subplot(224)
plt.imshow(img3, cmap='gray')
plt.title('IMG ESCALA DE GRISES')
plt.show()
```
